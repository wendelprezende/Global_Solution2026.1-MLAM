**3. CONSTRUÇÃO DE GRÁFICOS ESTATÍSTICOS EM PYTHON**

Gráfico 1 – Quantidade de Missões por Empresa

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("space_missions.csv", encoding="MacRoman")

top10 = df["Company"].value_counts().head(10)

plt.figure(figsize=(10,5))
top10.plot(kind="bar")

plt.title("Quantidade de Missões por Empresa")
plt.xlabel("Empresa")
plt.ylabel("Número de Missões")

plt.show()
```

Gráfico 2 – Distribuição dos Custos das Missões

```python
df = pd.read_csv("space_missions.csv", encoding="MacRoman")

df["Price"].dropna().hist(
   bins=8,
   figsize=(8,5)
)

plt.title("Distribuição dos Custos das Missões")
plt.xlabel("Preço (US$ milhões)")
plt.ylabel("Frequência")

plt.show()
```

**4. ANÁLISES UNIVARIADAS UTILIZANDO ESTATÍSTICA DESCRITIVA**

Código Python da Análise

```python
df = pd.read_csv("space_missions.csv", encoding="MacRoman")

df["Price"] = pd.to_numeric(df["Price"], errors='coerce')
price = df["Price"].dropna()
print("Média:", price.mean())
print("Mediana:", price.median())
print("Moda:", price.mode())

print("Máximo:", price.max())
print("Mínimo:", price.min())

print("Amplitude:",
     price.max() - price.min())

print("Variância:",
     price.var())

print("Desvio Padrão:",
     price.std())

print("Quartis:")
print(price.quantile([0.25, 0.50, 0.75]))
```
